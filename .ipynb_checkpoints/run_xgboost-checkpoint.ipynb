```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBRegressor

from Dataloader import Dataloader

dl = Dataloader()
train = dl.get_train()
print(train[0])
true_vals = train[:,-7]
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.plot(true_vals,c="b")

fig1.show()
```

```python
trainX, trainy = train[:, :-7], train[:, -1]
model = XGBRegressor(objective='reg:squarederror', n_estimators=1000)
model.fit(trainX, trainy)
```

```python
test = dl.get_test()
print(test[0,:-1].shape)
#model.predict(test[0,:-1])


true_vals = test[0:100,-1]

preds = np.zeros(len(true_vals))
for i in range(0, len(preds)):
    row = test[i,:-7].flatten()
    preds[i] = model.predict(np.asarray([row]))[0]

```

```python
fig1 = plt.figure(figsize=(18, 6))
ax1 = fig1.add_subplot(111)
ax1.plot(preds,c="r")
ax1.plot(true_vals,c="b")

fig1.show()
```
